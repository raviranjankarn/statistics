```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
samples = np.random.normal(size=10000) #Taking 10000 samples where they are normally distributed
max = np.max(samples) #max
min = np.min(samples) #max
range = max-min
mean = np.mean(samples) #mean
median = np.median(samples)#median
var = np.var(samples) #varience
std = np.std(samples)#Standard deviation
q1 = np.percentile(samples, 25) #First quartile
q3 = np.percentile(samples, 75) #Third quartile
```

```python
print("The number of samples are {}".format(len(samples)))
print("The max value of the samples is {}".format(max))
print("The min of the samples is {}".format(min))
print("The range of the samples is {}".format(range))
print("The mean of the samples is {}".format(mean))
print("The median of the samples is {}".format(median))
print("The variance of the samples is {}".format(var))
print("The standard deviation of the samples is {}".format(std))
print("The first Quartile of samples is {}".format(q1))
print("The third Quartile of samples is {}".format(q3))
print("The interquartile range (q3-q1) of the samples is {}".format(q3-q1))
```

```python
plt.figure(figsize=(20,10))
plt.hist(samples, bins=20,color='gray')
plt.axvline(x=mean, c='green', label='Mean')
plt.axvline(x=median, c='black', label='Median')
plt.axvline(x=mean-std, c='lime', label='1Std')
plt.axvline(x=mean+std, c='lime')
plt.axvline(x=mean-2*std, c='yellow', label='2Std')
plt.axvline(x=mean+2*std, c='yellow')
plt.axvline(x=mean-3*std, c='purple', label='3Std')
plt.axvline(x=mean+3*std, c='purple')
plt.axvline(x=q1, c='pink', label='Interquartile')
plt.axvline(x=q3, c='pink')
plt.xticks([-4,mean,median,mean-std,mean+std,mean-2*std,mean+2*std,mean-3*std,mean+3*std,q1,q3,4],rotation=75,fontsize= 8)
plt.legend()
plt.show()
```
